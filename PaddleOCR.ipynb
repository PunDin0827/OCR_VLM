```python
import os
import json
from pathlib import Path
from opencc import OpenCC
from paddleocr import PPStructureV3
```

```python
ocr = PPStructureV3(
    use_region_detection = False,
    use_table_recognition = True,
    use_formula_recognition = False,
    use_chart_recognition = False,
    use_doc_orientation_classify = True,
    use_doc_unwarping = False,
    use_textline_orientation = False,

    lang='chinese_cht',
    layout_detection_model_dir = "",
    text_detection_model_dir = "",
    text_recognition_model_dir = "",
    table_orientation_classify_model_dir = "",
    table_classification_model_dir = "",
    wireless_table_cells_detection_model_dir = "",
    wireless_table_structure_recognition_model_dir = "",
    wired_table_structure_recognition_model_dir = "",
    wired_table_cells_detection_model_dir = "",
    chart_recognition_model_dir = "",
    doc_orientation_classify_model_dir = "",
    textline_orientation_model_dir = ""
    )
```

```python
result = ocr.predict(input="",
    use_region_detection = False,
    use_table_recognition = True,
    use_formula_recognition = False,
    use_chart_recognition = False,
    use_doc_orientation_classify = True,
    use_doc_unwarping = False,
    use_textline_orientation = False,)

for res in result:
    res.print()
    res.save_to_json(save_path="")
    res.save_to_markdown(save_path="")
result_list = ocr.predict(input = "")
```

```python
res0 = result_list[0]
res0 = res0.json["res"]["table_res_list"]
cell_box = res0[0]["cell_box_list"]
html = res0[0]["pred_html"]

cell_box = json.dumps(cell_box, ensure_ascii=False, indent=2) # 表格
print(cell_box)

html = json.dumps(html, ensure_ascii=False, indent=2) # 位置訊息+文字
print(html)
```

```python
path = ""
os.makedirs(os.path.dirname(path) , exist_ok=True)
with open(path , "w" ,encoding = "utf-8") as f:
    f.write(cell_box)

path = ""
os.makedirs(os.path.dirname(path) , exist_ok=True)
with open(path , "w" ,encoding = "utf-8") as f:
    f.write(html)
```
